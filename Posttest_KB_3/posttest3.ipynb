```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("diabetes_prediction_dataset.csv")
df.info()
```

GRAFIK BATANG

Disini terlihat orang yang tidak memiliki diabetes itu kebanyakan berasal dari umur 1-40 tahun sedangkan yang memiliki diabetes itu berasal dari 1-60 tahun keatas.

```python
x = list(df.iloc[:,1])
y = list(df.iloc[:,1])
```

```python
sns.barplot(
    data = df,
    x    = 'diabetes', 
    y    = 'age',
)
plt.title('Grafik orang yang Memiliki Diabetes berdasarkan Umur')
plt.show()
```

SCATTER PLOT

Pada Scatter Plot Disini kebanyakan yang memiliki diabetes itu berasal dari indeks massa tubuh 1-60 keatas.

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10,10))
sns.scatterplot(
    data  = df,
    x     = 'diabetes',  
    y     = 'bmi'
)
plt.title('Orang yang memiliki diabetes berdasarkan indeks massa tubuh')
plt.show()
```

PIE CHART

Pada Pie Chart Dari data ini terlihat orang yang memiliki diabetes hanya 8,5% dari 10,000 orang dan yang tidak memiliki diabetes sebesar 91,5%

```python
df = df.dropna(subset=['diabetes'])
data_pie = df['diabetes'].value_counts().rename_axis('diabetes').reset_index(name='age')
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_pie.age, 
    labels     = data_pie.diabetes, 
    startangle = 90,
    autopct    = '%.1f%%'
)
plt.title('Data Orang Yang Memiliki Diabetes')
plt.show()
```

GRAFIK BOX PLOT

Pada Grafik Box Plot Disini keterangan 0 itu tidak ada dan 1 adalah ada dan dari data ini kebanyakan yang memiliki diabetes berasal dari umur 50 - 70 tahun keatas.

```python
plt.figure(figsize=(10,20))
sns.boxplot(
    data   = df,
    x      = 'diabetes',
    y      = 'age',
)

plt.xlabel('diabetes')
plt.ylabel('age')
plt.title('Orang yang Memiliki Diabetes berdasarkan umur')
plt.show()
```

HEATMAP

Pada heatmap ini Korelasi Antara Atribut Numerik Dan Diabetes ditampilkan untuk melihat hubungan atau keterkaitan antara atribut numerik dengan diabetes.

```python
atribut_numerik = ['age', 'HbA1c_level','blood_glucose_level','diabetes', 'hypertension', 'heart_disease']
```

```python

correlation_matrix = df[atribut_numerik].corr()
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Atribut Numerik dan diabetes')
plt.show()
```
