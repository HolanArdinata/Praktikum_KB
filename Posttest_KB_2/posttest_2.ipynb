3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya.

```python
# Jumlah record
jumlah_record = len(df)
# Jumlah attribute tipe objek
objek = len(df.dtypes[df.dtypes == 'object'])
# Nama attribute
atr = "\n".join(df.columns)
# Jumlah attribute tipe angka
angka = len(df.dtypes[df.dtypes == 'int64'])

# Membuat table
table = PrettyTable()
table.field_names = ["Attribute", "Value"]
table.add_row(["Jumlah Record", jumlah_record])
table.add_row(["Jumlah Attribute Tipe Object", objek])
table.add_row(["Nama Attribute", atr])
table.add_row(["Jumlah Attribute Tipe Angka", angka])


# Memprint Table
print(table)
```

4. Tampilkan nilai dataset pada baris dimulai dengan baris 103 hingga data terakhir dan kolom ke 4 hingga kolom terakhir menggunakan ke 2 metode slice.

```python
# metode .iloc (indeks berbasis posisi)
subset_iloc = df.iloc[103:, 3:]  

# Menampilkan hasil
print("Menampilkan hasil :")
print(subset_iloc)
```

```python
# metode .loc 
subset_loc = df.loc[103:, df.columns[3]:]  # Mulai dari baris 103 dan dari kolom ke-4

# Menampilkan hasil
print("\nMenampilkan Hasil :")
print(subset_loc)
```

5. Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).


```python
object_attribute = df.select_dtypes(include=['object'])
for attribute in object_attribute.columns:
    jumlah_label = df[attribute].nunique()
    print(f"Attribute '{attribute}' mempunyai {jumlah_label} label.")
```
